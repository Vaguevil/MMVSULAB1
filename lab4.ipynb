```
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def euler(dydx, x, y0, n):
    y = [y0]
    for i in range(n - 1):
        xi = x[i]
        yi = y[i]
        yi1 = yi + h * dydx(yi, xi)
        y.append(yi1)
    return x, y

def dydx(x, y):
    k = 0.42
    return k * y

x0, y0 = 0, 1
h = 0.1

x = np.arange(0, 10 + h, h) 
t = np.linspace(0, 10, 100)
x, y = euler(dydx, y0 = 1, x = x, n = len(x - 1))

solve = odeint(dydx, y0 = 1, t = t)
solve = solve.flatten()



plt.plot(x, y, label = r"euler")
plt.plot(t, solve, label = r"$x' = k(x)*x$")
plt.grid(True)
plt.legend()
plt.show()
```
