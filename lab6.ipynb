```
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np

x=np.linspace(0,1,100)
y=x+x*np.random.random(len(x))
A=np.vstack([x,np.ones(len(x))]).T
a,b=np.linalg.lstsq(A,y,rcond=None)[0]

plt.plot(x,y,'.')
plt.plot(x,x*a+b,'r')
plt.grid(True)
plt.show()
```
